```python
import time
import matplotlib.pyplot as plt
from tabulate import tabulate
```

Data barang dan stok

```python
# List merek barang dan stok barang yang sudah ditentukan
merek_barang = [
    "Aqua", "Ultra Milk", "Indomie", "Rinso", "Pepsodent",
    "Lifebuoy", "Mie Sedaap", "Sunsilk", "Frisian Flag", "Bear Brand",
    "Teh Pucuk", "Cimory", "Nutrisari", "Good Day", "Yakult",
    "So Klin", "Daia", "Sedaap Sambal", "Torabika", "Energen",
    "Downy", "Molto", "Dove", "Rexona", "Clear",
    "Pantene", "Lux", "Sanqua", "ABC Syrup", "Kecap Bango",
    "Gulaku", "Mayora", "Roma Biscuit", "Silver Queen", "Milo",
    "Nescafe", "Indocafe", "Nestle Pure Life", "Oreo", "Tango",
    "KitKat", "Aqua Reflections", "Pocari Sweat", "You C1000", "Sprite",
    "Coca Cola", "Fanta", "Pepsi", "Teh Kotak", "Frestea"
]

stok_barang = [
    500, 300, 1000, 150, 250,
    800, 600, 750, 400, 450,
    700, 100, 600, 500, 650,
    900, 200, 800, 350, 300,
    750, 550, 650, 700, 850,
    400, 100, 550, 600, 950,
    720, 780, 800, 660, 750,
    470, 580, 620, 890, 650,
    810, 950, 700, 630, 720,
    820, 400, 750, 860, 900
]
```

Mencari barang dengan stok terbanyak dengan cara iteratif

```python
# Algoritma Iteratif
def cari_stok_iteratif(merek_barang, stok_barang):
    stok_terbanyak = stok_barang[0]
    indeks_terbanyak = 0

    for i in range(1, len(stok_barang)):
        if stok_barang[i] > stok_terbanyak:
            stok_terbanyak = stok_barang[i]
            indeks_terbanyak = i

    return merek_barang[indeks_terbanyak], stok_terbanyak
```

Mencari barang dengan stok terbanyak dengan cara rekursif

```python
# Algoritma Rekursif
def cari_stok_rekursif(merek_barang, stok_barang, indeks, stok_terbanyak, indeks_terbanyak):
    if indeks >= len(stok_barang):
        return indeks_terbanyak, stok_terbanyak
    if stok_barang[indeks] > stok_terbanyak:
        stok_terbanyak = stok_barang[indeks]
        indeks_terbanyak = indeks
    return cari_stok_rekursif(merek_barang, stok_barang, indeks + 1, stok_terbanyak, indeks_terbanyak)
```

Analisis waktu eksekusi

```python
# Analisis Running Time
sizes = [5, 10, 15, 20, 25]
iterative_times = []
recursive_times = []

for size in sizes:
    subset_merek_barang = merek_barang[:size]
    subset_stok_barang = stok_barang[:size]

    # Uji iteratif
    start_time = time.time()
    cari_stok_iteratif(subset_merek_barang, subset_stok_barang)
    iterative_times.append(time.time() - start_time)

    # Uji rekursif
    start_time = time.time()
    cari_stok_rekursif(subset_merek_barang, subset_stok_barang, 0, subset_stok_barang[0], 0)
    recursive_times.append(time.time() - start_time)
```

Membandingkan performa kedua algoritma

```python
# Membuat tabel perbandingan
data_tabel = []
for i, size in enumerate(sizes):
    data_tabel.append([size, recursive_times[i], iterative_times[i]])

# Format tabel dengan tabulate
headers = ["n", "Recursive Time (s)", "Iterative Time (s)"]
print(tabulate(data_tabel, headers=headers, tablefmt="grid"))

# Plot hasil
plt.figure(figsize=(10, 6))
plt.plot(sizes, iterative_times, label="Iteratif", marker="o")
plt.plot(sizes, recursive_times, label="Rekursif", marker="x")
plt.title("Perbandingan Running Time")
plt.xlabel("Ukuran Data (n)")
plt.ylabel("Waktu Eksekusi (detik)")
plt.legend()
plt.grid()
plt.show()
```

Hasil pencarian barang dengan stok terbanyak

```python
# Eksekusi dengan data lengkap
merk_iteratif, stok_iteratif = cari_stok_iteratif(merek_barang, stok_barang)
print(f"Iteratif: Jenis barang dengan stok terbanyak: {merk_iteratif} dengan jumlah stok: {stok_iteratif}")

indeks_rekursif, stok_rekursif = cari_stok_rekursif(merek_barang, stok_barang, 0, stok_barang[0], 0)
print(f"Rekursif: Jenis barang dengan stok terbanyak: {merek_barang[indeks_rekursif]} dengan jumlah stok: {stok_rekursif}")
```
